Topic 13 — Advanced Functions

```python
'''Problem 1 — Sum Using *args

Create function that returns sum of unlimited numbers.'''
def unlimited_num_sum(*args):
    sum = 0
    for i in args:
        sum += i
        return sum
    
print(unlimited_num_sum(1,2,3,4,5,67,8,9,0))
```

```python
'''Problem 2 — Find Maximum

Accept unlimited numbers and return maximum value.'''

def unlimited_num_max(*args):
    max_num = max(args)
    return max_num

print(unlimited_num_max(1,22,33,454,654))
```

```python
'''Problem 3 — Student Profile (**kwargs)

Function prints all student details passed dynamically.

Example call:

student(name="Ali", age=20, course="AI")'''
def student(**kwargs):
    for k, v in kwargs.items():
        print(k, ":", v)

student(name="Ali", age=20, course="AI")
```

```python
'''
Problem 4 — Mixed Arguments

Create function:

report(title, *scores, **details)

Print:

title

average score

extra details'''

def report(title, *scores, **details):
    avg = sum(scores) / len(scores)
    print("Title:", title)
    print("Average:", avg)
    print("Details:", details)

report("Exam", 80,90,70, name="Hamza")
```

```python
'''
Problem 5 — List Unpacking

Given:

values = [5,10,15]

Pass into function using unpacking.'''

values = [5,10,15]

def add(a,b,c):
    print(a+b+c)

add(*values)
```

Topic 14 — Lambda Functions (Anonymous Functions)


```python
'''Problem 1 — Square Using Lambda

Create lambda to square a number.'''
square = lambda x : x**2
square(5)
```

```python
'''Problem 2 — Add Three Numbers

Lambda that adds three numbers.'''
add = lambda x,y,z : x + y + z
add(1,2,3)
```

```python
'''Problem 3 — Filter Odd Numbers

From list:

nums = [1,2,3,4,5,6,7]

Return only odd numbers using lambda.'''

nums = [1,2,3,4,5,6,7]
odds = list(filter(lambda x : x%2 !=0, nums))
print(odds)
```

```python
'''Problem 4 — Sort by Last Character
words = ["apple", "banana", "kiwi"]

Sort using last character.'''

words = ["apple", "banana", "kiwi"]

sorted_words = sorted(words, key=lambda x: x[-1])
print(sorted_words)
```

```python
'''Problem 5 — Double List Values

Use map() + lambda to double numbers.'''
nums = [1,2,3,4]

result = list(map(lambda x: x*2, nums))

print(result)
```
